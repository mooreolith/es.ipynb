### Vega Lite Output
The Vega Lite instructions are little terse, but discernible with a little experimentation. It's important to pay close attention to spelling and attributes, because small oversights in reading can mean the difference between a working graphic, and no working graphic.


```ecmascript
var data = [
  {"x": Math.random(), "y": Math.random()},
  {"x": Math.random(), "y": Math.random()},
  {"x": Math.random(), "y": Math.random()},
  {"x": Math.random(), "y": Math.random()},
  {"x": Math.random(), "y": Math.random()},
  {"x": Math.random(), "y": Math.random()}
];

var spec = {
  "$schema": "https://vega.github.io/schema/vega-lite/v4.json",
  "data": {"values": data},
  "mark": "circle",
  "encoding": {
    "x": {"field": "x", "type": "quantitative"},
    "y": {"field": "y", "type": "quantitative"}
  }
}

spec
```

```ecmascript
var width = 650;
var height = 500;

var scene = new THREE.Scene();
var camera = new THREE.PerspectiveCamera( 75, width / height, 0.1, 1000 );

var renderer = new THREE.WebGLRenderer();
renderer.setSize( width, height );
document.querySelector('#renderTarget').appendChild( renderer.domElement );

var geometry = new THREE.BoxGeometry( 1, 1, 1 );
var material = new THREE.MeshBasicMaterial( { color: 0x00ff00 } );
var cube = new THREE.Mesh( geometry, material );
scene.add( cube );

camera.position.z = 5;

function animate() {
  requestAnimationFrame( animate );

  cube.rotation.x += 0.01;
  cube.rotation.y += 0.01;

  renderer.render( scene, camera );
};

animate();
var result = 'rendering';
result
```
